```python
import urllib.request

def extract_data (url, filename):
  try:
    urllib.request.urlretrieve(url, filename)

  except Exception as e:
    print(e)
```

```python
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/2be28d90-3a90-4af1-a3f6-f28c1e25880a/download/tmp7602cia8.csv", "data\dados_2018")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/ea2e4696-4a2d-429c-9807-d02eb92e0222/download/tmpcje3ep_w.csv", "data\dados_2019")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/6ff6a6fd-3141-4440-a880-6f60a37fe789/download/tmpcv_10m2s.csv", "data\dados_2020")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/f53ebccd-bc61-49f9-83db-625f209c95f5/download/tmp88p9g82n.csv", "data\dados_2021")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/81a7b022-f8fc-4da5-80e4-b160058ca207/download/tmpfm8veglw.csv", "data\dados_2022")
extract_data("https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/e6013a93-1321-4f2a-bf91-8d8a02f1e62f/download/tmp182oee02.csv", "data\dados_2023")
```

```python
arquivos = [
    "data\dados_2018",
    "data\dados_2019",
    "data\dados_2020",
    "data\dados_2021",
    "data\dados_2022",
    "data\dados_2023"
]
```

```python
dfs = {}
```

```python
import pandas as pd

for arquivo in arquivos:
  ano = arquivo.split("_")[-1].split(".")[0]
  dfs[ano] = pd.read_csv(arquivo)
```

```python
import boto3
```

```python
aws_access_key_id = ""
aws_secret_access_key = ""
aws_session_token = ""

region_name = "us-east-1"

boto3.setup_default_session(
    aws_access_key_id = aws_access_key_id,
    aws_secret_access_key = aws_secret_access_key,
    aws_session_token = aws_session_token,
    region_name = region_name
)

s3 = boto3.client("s3")
```

```python
content = """
Olá, S3
"""

with open ("hello-s3.txt", "w+") as f:
  f.write(content)
```

```python
s3.upload_file("hello-s3.txt", "datalake-pipeline-lucasts", "bronze/hello-s3")
```

```python
from io import BytesIO

for ano, df in dfs.items():
  parquet_buffer = BytesIO()
  df.to_parquet(parquet_buffer)

  s3.put_object(
      Bucket = "datalake-pipeline-lucasts",
      Key = f"bronze/dados_{ano}.parquet",
      Body = parquet_buffer.getvalue()
  )
```

```python
response = s3.list_objects(Bucket="datalake-pipeline-lucasts")
```

```python
keys = [obj["Key"] for obj in response ["Contents"]]
print(keys)
```
